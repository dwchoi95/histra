```python
import ast

class CommentAdder(ast.NodeTransformer):
    def visit_FunctionDef(self, node):
        # 1. 삽입할 주석(독스트링)용 AST 노드 생성
        comment_node = ast.Expr(
            value=ast.Constant(value="자동으로 삽입된 주석입니다.")
        )
        
        # 2. 함수의 기존 본문 맨 앞에 주석 추가
        node.body.insert(0, comment_node)
        
        # 3. 변경된 노드 반환
        return node

# 테스트용 소스 코드
code = """
def my_function():
    print("Hello, World!")
"""

# 1. AST 파싱
tree = ast.parse(code)

# 2. NodeTransformer 적용
transformer = CommentAdder()
transformed_tree = transformer.visit(tree)

# 3. 변경된 AST를 파이썬 코드로 다시 변환 
# (Python 3.9 이상에서 ast.unparse 사용 가능)
print(ast.unparse(transformed_tree))
```
